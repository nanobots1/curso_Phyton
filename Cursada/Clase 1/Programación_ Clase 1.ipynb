## Variables

Las variables se componen de un "nombre" y de un "valor". Python distingue entre mayusculas y minusculas

```python
# variable se llama "nombre" y el valor es 'Christian' (string: cadena de caracteres)
nombre = 'Christian'
print(nombre)
Nombre = "Chris"
print(Nombre)

# otra variable de tipo string
print()
edad = "26 con string"
print(edad)

# variable se llama "edad" y el valor es 26 (entero)
edad = 26
print(edad)

# variable se llama "peso" y el valor es 64.9 (flotante)
peso = 64.9

```

```python
# el codigo se ejecuta de arriba hacia abajo.
# si se usan las variables del mismo nombre con nuevos valores se actualizan.
edad = "26 con string"
edad = 26

print(edad)
```

## Tipo de datos: Operaciones aritmeticas

```python
calculo = 7 + 4
print("esto es el resultado de numeros enteros: ", calculo)
print()
calculo = "7" + "4" # concatenación
print("esto es el resultado de dos cadenas de caracteres: ", calculo)
print()
calculo = 7 * "4"
print("esto es el resultado de un entero y un caracter: ", calculo)
```

Condicionales: if, else

```python
edad = input("Ingrese su edad: ")

# No se pueden realizar comparaciones con diferentes tipos de datos
if edad >= 18:
  print("Eres mayor de edad")
else:
  print("Eres menor de edad")
```

```python
edad = int(input("Ingrese su edad: "))

if edad >= 18:
  print("Eres mayor de edad")
else:
  print("Eres menor de edad")
```

if, elif, else

```python
edad = int(input("Ingrese su edad: "))

if edad >= 0 or edad < 3:
  print("Eres un bebe")
else:
  print("Eres mayor de edad")
```

while

```python
contador = 0

while contador < 5:
  print("contador es igual a: ", contador)
  contador += 1
```

for

```python
frutas = ["manzana", "banana", "pera", "durazno"]

for fruta in frutas:
  if fruta == "uva":
    break
  print(fruta)
```

## Ejercicio 3

```python
# calificaciones es una lista de tuplas
calificaciones = [("Alice", 8), ("Bob", 9), ("Charlie", 7), ("David", 6), ("Eva", 4)]

# iniciar variables
suma_calificaciones = 0
numero_estudiantes = len(calificaciones)

# sumar todas las calificaciones
for estudiante, calificacion in calificaciones:
  # suma_calificaciones = suma_calificaciones + calificacion
  suma_calificaciones += calificacion

# calcular el promedio
promedio = suma_calificaciones / numero_estudiantes

# mostrar el resultado del promedio
print("El promedio de calificaciones de los estudiantes es de: ", promedio)
```
