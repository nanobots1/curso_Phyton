# Ejercicio 1

Enunciado:

Haga un programa que lea por consola cadenas de caracteres y que
por cada `string` leído imprima la cantidad de caracteres del mismo. El
programa termina al ingresar la cadena vacía. Al finalizar informar la
longitud promedio de las palabras ingresadas.

Ayuda: si `st` es una variable de tipo string, `len(st)` devuelve la cantidad
de caracteres del `string`.

```python
# línea separatoria
dash_line = '-'.join('' for x in range(10))

# inicializamos variables
contador = 0
suma = 0

# solicitamos al usuario que ingrese un string por teclado y lo almacenamos en la variable "palabra"
palabra = input("Ingrese un string por favor: ")

# comienza el bucle while y termina cuando se ingresa una cadena vacía ""
while palabra != "":
  # incrementamos el contador en cada lectura de string
  contador = contador + 1
  # obtenemos la cantidad de caracteres
  longitud = len(palabra)
  # lo mostramos por pantalla
  print(f"El string '{palabra}' tiene {longitud} caracteres")
  # incrementamos la suma
  suma = suma + longitud
  # actualizamos la variable palabra
  palabra = input("Ingrese un string por favor: ")

print(dash_line)
promedio = suma / contador
print("El promedio de las logitudes es: ", promedio)
```

# Ejercicio 2

Enunciado:

Escriba un programa que multiplique por `365` el número entero ingresado por el usuario desde la consola. El resultado debe ser impreso en la consola dígito por dígito, uno en cada línea, comenzando por el dígito menos significativo al más significativo.

Ayuda: convertir el número obtenido a `string` y utilizar operaciones de `string`

```python
# solicitamos que se ingrese un año
numero = int(input("Ingrese un número entero: ")) * 365

# convertimos el número a cadena (string)
cadena = str(numero)

# iniciamos variables para recorrer la cadena
desde = len(cadena) - 1
hasta = -1
paso = -1

# recorremos la cadena desde el dígito menos significativo al más significativo (de derecha a izquierda)
for indice in range(desde, hasta, paso):
  print(cadena[indice])
```

# Ejercicio 3

Enunciado:

Escriba un programa que lea del teclado una secuencia de números
que finaliza con un 0 e imprima el menor y el mayor de los números
ingresados. Al finalizar el proceso imprimir la lista de valores
comprendidos entre el máximo y mínimo valor.

Ayuda: probar con `listas` y la función `sort()`

```python
# inicializamos la lista vacia
lista_numeros = []

while True:
  # solicitamos un numero
  numero = int(input("Ingrese numero: "))
  # evaluamos si es igual a '0'
  if numero == 0:
    # termina el bucle
    break
  # si no es '0', agregamos el número a la lista
  lista_numeros.append(numero)

# ordenamos la lista de manera inversa usando la función sort
lista_numeros.sort(reverse=True)


# mostramos el menor y mayor número
print(dash_line)
print("El menor número es: ", lista_numeros[-1])
print("El mayor número es: ", lista_numeros[0])

# mostramos la lista completa
print(dash_line)
print("La lista de número es: ", lista_numeros)
```

# Ejercicio 4

Enunciado:

Escriba un programa que lea 2 palabras y determine si son
palíndromos entreellas. (Ej: “abbccd” y “dccbba” son palíndromos).

a) Resuélvalo recorriendo cada palabra y comparándolas

b) Resuélvalo usando funciones predefinidas de lista

```python
# Inreso de palabras. Convertir a minúsculas para ser insensible a mayúsculas/minúsculas
palabra1 = input("Ingrese primera palabra: ").lower()
palabra2 = input("Ingrese segunda palabra: ").lower()
```

a) recorriendo cada palabra y comparandolas

```python
# es_palindromo nos indicara si las palabras son palindromo
es_palindromo = True

# el indice es para recorrer en sentido inverso la segunda palabra
indice = -1

# recorremos la primer palabra (caracter a caracter)
for letra1 in palabra1:
  # comparamos los caracteres de la primer palabra
  # (de izquierda a derecha) y la segunda palabra
  # (sentido reverso). Si no son iguales,
  # las palabras no son palindromos.
  if letra1 != palabra2[indice]:
    # cambiamos el valor booleano
    es_palindromo = False
    # terminamos el bucle for
    break
  else:
    # si los caracteres son iguales, seguimos recorriendo las palabras
    indice = indice - 1

# imprimimos los resultados
if es_palindromo:
    print("Ambas palabras son palíndromos")
else:
    print("No son palíndromos")
```

b) usando funciones predefinidas de lista

```python
# utilizar el operador slicing [::-1] para invertir un string
palabra2_invertida = palabra2[::-1]

if palabra1 == palabra2_invertida:
    print("Son palíndromas entre sí.")
else:
    print("No son palíndromas.")
```

# Ejercicio 5

Enunciado:

Implemente un programa que ingrese la marca de cada producto vendido
en un kiosco e imprima al final del día un listado que indique para cada
marca la cantidad total de productos vendidos de la misma.
Además imprima la cantidad total de marcas diferentes
ingresadas. Resolver usando diccionario `dict()` y conjunto `set()`.

```python
# inicializamos las variables de diccionario y conjunto
diccionario_marcas = dict()
unicas_marcas = set()

# iniciamos bucle while
while True:
  # solicitamos la marca
  marca = input("Ingrese marca. Para salir EXIT: ")
  # finalizamos el bucle si marca es igual a "EXIT"
  if marca == "EXIT":
      break
  # guardamos las marcas en cada iteración (solo únicas)
  unicas_marcas.add(marca)
  # guardamos la cantidad vendida
  cantidad = int(input("Ingrese cantidad: "))

  # preguntamos si la marca se encuentra en el diccionario
  if marca in diccionario_marcas:
    # le actualizamos la cantidad a la marca
    diccionario_marcas[marca] += cantidad
  else:
    # le agregamos la cantidad a la marca
    diccionario_marcas[marca] = cantidad

print(dash_line)
print("La lista de marcas únicas es: ", unicas_marcas)
print(dash_line)
print("El listado de productos por marcas es: ", diccionario_marcas)
```

# Ejercicio 6

Enunciado:

Escriba un programa que pida un año por pantalla y diga si es
`bisiesto`. Un año es `bisiesto` si es divisible por `4` pero no por `100`. Si es
divisible por `100`, para ser `bisiesto` debe ser divisible por `400`.

```python
# solicitamos el año por teclado
anio = int(input("Ingrese un año: "))

# Evaluamos todos los casos:
# Paso 1: Si es divisible por 4, ir al paso 2. Caso contrario, no es bisiesto.
# Paso 2: Si es divisible por 100, ir al paso 3. Caso contrario, no es bisiesto.
# Paso 3: Si es divisible por 400 es bisiesto. Caso contrario, no es bisiesto.
if anio % 4 == 0 and (anio % 100 != 0 or anio % 400 == 0):
    print(anio," es bisiesto")
else:
    print(anio," NO es bisiesto")
```
