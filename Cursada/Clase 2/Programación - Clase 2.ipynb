```python
# calificaciones es una lista de tuplas
calificaciones = [("Alice", 8), ("Bob", 9), ("Charlie", 7), ("David", 6), ("Eva", 4)]

# iniciar variables
suma_calificaciones = 0
numero_estudiantes = len(calificaciones)

# sumar todas las calificaciones
for estudiante, calificacion in calificaciones:
  # suma_calificaciones = suma_calificaciones + calificacion
  suma_calificaciones += calificacion

# calcular el promedio
promedio = suma_calificaciones / numero_estudiantes

# mostrar el resultado del promedio
print("El promedio de calificaciones de los estudiantes es de: ", promedio)
```

```python
def calcular_promedio(lista, suma, estudiantes):
  # sumar todas las calificaciones
  for estudiante, cal in lista:
    suma += cal

  # calcular el promedio
  promedio = suma / estudiantes

  return "El promedio de calificaciones de los estudiantes es de: ", promedio
```

```python
def precio_promedio(lista):
  suma_precios = sum(lista)
  promedio = suma_precios / len(lista)
  return promedio

lista_precios = [10, 20, 50, 80, 60, 40]
promedio = precio_promedio(lista_precios)

print("El promedio de la lista de precios es: ", promedio)
```

Recursi√≥n

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

factorial(30)
```

Ejercicio 2

```python
def suma_recursiva(n):
  if n == 1:
    return 1
  else:
    return n + suma_recursiva(n - 1)

suma_recursiva(5)
```

```python
def division_segura(dividendo, divisor):
  try:
    resultado = dividendo / divisor
    return resultado
  except ZeroDivisionError:
    return "Error: no es posible dividir por cero (0)."

division_segura(10, 0)
```
