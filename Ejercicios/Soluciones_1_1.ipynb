Enunciado:

Haga un programa que lea por consola cadenas de caracteres y que
por cada `string` leído imprima la cantidad de caracteres del mismo. El
programa termina al ingresar la cadena vacía. Al finalizar informar la
longitud promedio de las palabras ingresadas.

Ayuda: si `st` es una variable de tipo string, `len(st)` devuelve la cantidad
de caracteres del `string`.

```python
# separador de lineas
separador = '-'.join('' for x in range(20))

#inicializo variables
cantidad = 0
longitud = 0

#ingreso dato
palabra = input("Ingrese palabra : ")

# bucle mientras que
while palabra != "" :
  #acumulo cantidad palabras ingresadas
  cantidad += 1
  #acumulo longitud larga palabra
  longitud += len(palabra)
  # ingreso nuevo dato
  palabra = input("Ingrese palabra : ")

# calculo promedio
promedio = longitud / cantidad
#imprimo datos
print (separador)


print(f"promedio de longitud de palabras, {promedio}")
```
